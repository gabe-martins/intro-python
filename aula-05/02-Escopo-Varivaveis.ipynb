# **Escopo de Variaveis**

O uso de escopo de variáveis é importante para garantir a **organização** e **controle do código** em um programa. O escopo se* refere à visibilidade de uma variável dentro do programa, ou seja, em que partes do código a variável pode ser acessada.*

Ao definir o escopo de uma variável, é possível evitar **conflitos** com outras variáveis de mesmo nome em diferentes partes do programa e também garantir que uma variável **não seja acessada indevidamente em locais não autorizados**.

Além disso, o uso adequado de escopo pode ajudar na **otimização** do uso de memória, já que variáveis definidas em escopos mais restritos podem ser **destruídas** automaticamente quando o escopo terminar, liberando recursos.

Em resumo, o uso de escopo de variáveis é importante para garantir a **organização, segurança e eficiência** do código em um programa.



```python
def teste():
  x = 8 
  print(f'x = {x}')
  print(f'n = {n} na função')

n = 2
print(f'n = {n}')
teste()
```

```python
def teste(b):
  b += 4
  return b

a = 5
b = 2
print(f'B da função: {teste(a)}')
print(f'B Global {b}')
```

```python
def funcao():
  n1 = 4
  print(f'N1 dentro {n1}')

n1 = 2
print(f'N1 global {n1}')
funcao()
```

# **Retorno de variaveis**

```python
def somar(a=0,b=0):
  s = a + b
  return s

r1 = somar(5,9)
r2 = somar(8, 17)
r3 = somar(7)
```

```python
print(f'soma = {r1}')
print(f'Area é = {r2}')
print(f'seu produto custa R$ {r3}')
```
