# Execicio 01 

1) Crie um programa que vai ler do teclado **vários número** e colocar em uma **lista**.

## resolução 1

```python
numeros = []
```

```python
for i in range(5):
  numero = input("Digite um número (ou pressione Enter para encerrar): ")
  numeros.append(int(numero))
```

## resolução 2

```python
qtde = int(input('Quantos numero deseja inserir? '))

for i in range(qtde):
  numero = input("Digite um número (ou pressione Enter para encerrar): ")
  numeros.append(int(numero))
```

## resolução 3

```python
while True:
    numero = input("Digite um número (ou pressione Enter para encerrar): ")

    # 0
    # Null == None
    # Vazio == ""
    
    if numero == "":
        break

    numeros.append(int(numero))

print("Números lidos:", numeros)
```

A) **Quantos** números foram digitados?

```python
print(f'Foram digitados {len(numeros)} numeros')
```

B) **Mostrar** a lista de valores em ordem.

```python
numeros.sort()
print(numeros)
```

C) Mostrar se o valor **5** foi ao não digitado na lista.

```python
for i in numeros:
  if i == 5:
    print(i)

if 5 not in numeros:
  print('Não digitou o numero 5')
```
