```python
filmes = []

filme1 = {'titulo' : 'Star Wars','ano' : 1977, 'diretor': 'George Lucas'}
filme2 = {'titulo' : 'Titanic','ano' : 1998,'diretor': 'James Camerom'}

filmes.append(filme2)
filmes.append(filme1)

filmes
```

```python
filme1.keys()
```

```python
filme2.values()
```

```python
filmes[0].values()
```

```python
filme2.update({'ano': 1999})
filme2
```

```python
filme2['premios'] = 3
filme2
```

```python
for k, v in filme1.items():
  print(k, v) 
```

```python
for i in filmes:
  print(i)
  
  for k, v in i.items():
    print(v)  
```
