# LOOP FOR

```python
for i in range(10):
  print(f'passo {i}')
```

```python
for i in range(10):
  print(i)
```

```python
for x in range(5):
  print('x: ', x)
  for y in range(2):
    print(' y: ', y)
```

# LOOP WHILE

https://www.w3schools.com/python/python_while_loops.asp

```python
COUNT = 0

while COUNT <= 5:
  COUNT += 1
  print(COUNT)
  
```

```python
num = 10
while num >= 5:

  print(num)
  num -= 1

  if num == 6:
    break

print('Acabou')
```
