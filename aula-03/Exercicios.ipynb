# **Exercio 01**
Crie um programa em python que mostre os numeros pares que estão entre 1 e 30

```python
# Programa para mostrar números pares entre 1 e 30

# Loop para percorrer os números entre 1 e 30
for i in range(31):

  # Verifica se o número é par
  if i % 2 == 0:

    # Imprime o número par
    print(i, '- par')
    
  else:
    # Imprime o número impar
    print(i, '- impar')
```

```python
num = int(input('Digite um numero: '))

if num % 2 == 0:
  print(f'{num} é par')
else:
  print('impar')
```
