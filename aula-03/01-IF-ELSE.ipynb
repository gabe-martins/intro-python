```python
# Aluno
n1 = 6.0
n2 = 3.5

avg = (n1 + n2)/2

print(avg)

if avg >= 6.0:
  print('Aprovado!')
else:
  print('Reprovado!')
```

## And

```python
x = 0
y = 2
if (x == 0) and (y > 1):
  print('Condições corretas')
else:
  print('Condições errada')
```

## OR

```python
x = 0
y = 0
if (x == 0) or (y > 1):
  print('Condições corretas')
else:
  print('Condições errada')
```

```python
num = int(input('Numero: '))

if 1 >= num <= 10:
  print('True')
else:
  print('False')
```
