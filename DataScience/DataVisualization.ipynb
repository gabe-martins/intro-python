# Titanic

```python
# Carrega os dados
titanic = sns.load_dataset("titanic")

# Mostra os dados
titanic.head(2)
```

```python
titanic.loc[titanic['survived'] == 0, 'survived'] = 'no'
titanic.loc[titanic['survived'] == 1, 'survived'] = 'yes'
titanic.head(2)
```

```python
counts_survived = titanic['survived'].value_counts()
counts_survived
```

```python
# Plotly = px

px.pie(titanic, names=counts_survived.index, values=counts_survived.values, hole=.5, width=600, height=300)
```

```python
px.bar(titanic, x='sex', y='survived', width=800, height=400)
```

```python
px.scatter(titanic,x='sex', y=counts_survived, width=800, height=400)
```
